```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

import os

from brisk.visualization import plots_imu
from brisk.analysis import parameters
from brisk.utils import path
```

# Get all subjects

```python
subjects = path.get_subjects()
print('Subjects found:')
for s in subjects:
    print(f'\t{s}')
```

# Get parameters from all subjects

```python
all_data = []
glob = []
update = False
for s in subjects:
    dd = parameters.cycle_parameters(s, update)
    gg = parameters.global_parameters(s, update)
    dd['subject'] = s
    gg['subject'] = s
    all_data.append(dd)
    glob.append(gg)

all_data = pd.concat(all_data, axis=0, ignore_index=True)
glob = pd.concat(glob, axis=0, ignore_index=True)
all_data.head()
```

# Accelerometer

## Trunk

```python
plots_imu.plot_parameters(all_data, 'trunk', 'acc')
```

## Arm

```python
plots_imu.plot_parameters(all_data, 'arm', 'acc')
```

## Forearm

```python
plots_imu.plot_parameters(all_data, 'forearm', 'acc')
```
